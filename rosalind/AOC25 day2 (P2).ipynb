```python

answer_list = []
for number in number_list:
    length = len(str(number))
    for n in range( 1, length + 1):
        if n > length/2:
            continue
        if length % n == 0:
            if str(number)[:n] * (length//n) == str(number):
                answer_list.append(int(number))
answer = sum(list(set(answer_list)))
```
