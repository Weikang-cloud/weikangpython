```python
input_list = []
with open('list.txt', 'r') as file:
    for line in file:
        line = line.strip()
        input_list.append(line)
print(input_list[:10])
```

```python
 
start_position = 50
new_position = []
times_of_zero = 0
for dial in input_list:
    direction = dial[0]
    step = int(dial[1:])
    if direction == 'L':
        start_position = start_position - step
    elif direction == 'R':
        start_position = start_position + step
    if start_position < -100:
        start_position = 100 - int(str(start_position)[-2:])
    if start_position > 100:
        start_position = int(str(start_position)[-2:])
    if -100 < start_position < 0:
        start_position = 100 + start_position
    if 0 < start_position < 100:
        start_position = start_position
    else:
        start_position = 0
    
        
    
    new_position.append(start_position)
```
