```python
list_of_all_lists = []
with open('day6.txt') as file:
    for line in file:
        l = line.split()
        list_of_all_lists.append(l)
```

```python
result = 0
for position in range(len(list_of_all_lists[0])): 
    operator = list_of_all_lists[-1][position]
    if operator == '+':
        total = 0 
    elif operator == '*':
        total = 1
    for lists in list_of_all_lists[:-1]:
        current_number = lists[position]
        if operator == '+':
            total += int(current_number)
        elif operator == '*':
            total *= int(current_number)
    #print(total)
    result += total
```
